Ejercicio 1: Traslación

*   Cargar una imagen médica en escala de grises (puede ser una radiografía o resonancia magnética).
*   Aplicar una traslación de (50, 30) píxeles en la dirección (x, y).
*   Aplicar una traslación con valores decimales, por ejemplo, (20.5, 15.5) píxeles.
*   Mostrar la imagen original y las imágenes transformadas.





En este ejercicio, se aplicarán transformaciones de traslación a una imagen médica en escala de grises. Vamos a trasladar la imagen dos veces: una con un desplazamiento entero en píxeles y otra con un desplazamiento decimal. Mostraremos los resultados visualmente.

Para empezar, cargamos la imagen desde el almacenamiento de Google Drive. Es importante que la imagen esté en escala de grises para simplificar el procesamiento.
Definimos una matriz de traslación que moverá la imagen 50 píxeles a la derecha (eje X) y 30 píxeles hacia abajo (eje Y).
Aplicamos una traslación con valores decimales. Esto mueve la imagen de manera más precisa, pero es útil para algunas aplicaciones que requieren subpíxeles.

```python
# Paso 1: Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Paso 2: Importar librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Paso 3: Ruta a la imagen (ajusta si está dentro de una carpeta)
ruta_imagen = "/content/drive/My Drive/Imagen por Resonancia Magnética (IRM).png"

# Paso 4: Cargar la imagen en escala de grises
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

# Validar que se haya cargado correctamente
if imagen is None:
    raise ValueError("No se pudo cargar la imagen. Verifica el nombre o la ruta.")

# Dimensiones de la imagen
alto, ancho = imagen.shape

# Traslación 1: (50, 30)
M1 = np.float32([[1, 0, 50], [0, 1, 30]])
imagen_trasladada_1 = cv2.warpAffine(imagen, M1, (ancho, alto))

# Traslación 2: (20.5, 15.5)
M2 = np.float32([[1, 0, 20.5], [0, 1, 15.5]])
imagen_trasladada_2 = cv2.warpAffine(imagen, M2, (ancho, alto))

# Mostrar las imágenes
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagen_trasladada_1, cmap="gray")
plt.title("Traslación (50, 30)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagen_trasladada_2, cmap="gray")
plt.title("Traslación (20.5, 15.5)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

Ejercicio 2: Rotación
*   Aplicar una rotación de 45° alrededor del centro de la imagen.
*   Mostrar la imagen original y la imagen rotada.


En este ejercicio, aplicaremos una rotación de 45° sobre la imagen. La rotación se realizará alrededor del centro de la imagen, lo que es útil cuando no queremos que la imagen se desplace.

```python
from google.colab import drive
drive.mount('/content/drive')

# Importar librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ruta a la imagen (ajusta si está en una carpeta específica)
ruta_imagen = "/content/drive/My Drive/Imagen por Resonancia Magnética (IRM).png"

# Cargar la imagen en escala de grises
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

# Validar si la imagen se cargó correctamente
if imagen is None:
    raise ValueError("No se pudo cargar la imagen. Verifica la ruta o el nombre del archivo.")

# Dimensiones
alto, ancho = imagen.shape

# Definir la rotación: 45° alrededor del centro de la imagen
angulo = 45
centro = (ancho // 2, alto // 2)
escala = 1.0

# Matriz de rotación
M_rotacion = cv2.getRotationMatrix2D(centro, angulo, escala)

# Aplicar la transformación
imagen_rotada = cv2.warpAffine(imagen, M_rotacion, (ancho, alto))

# Mostrar imágenes
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(imagen_rotada, cmap="gray")
plt.title("Imagen Rotada 45°")
plt.axis("off")

plt.tight_layout()
plt.show()
```

Ejercicio 3: Escala
*   Aplicar un escalado al 150% de tamaño y otro al 50% de tamaño.
*   Mostrar la imagen original y los resultados de escalas.

En este ejercicio, escalaremos la imagen al 150% y al 50% de su tamaño original. Esto afecta la resolución de la imagen, aumentando o reduciendo el número de píxeles.

```python
from google.colab import drive
drive.mount('/content/drive')

# Librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ruta de la imagen desde tu Drive
ruta_imagen = "/content/drive/My Drive/Imagen por Resonancia Magnética (IRM).png"

# Cargar la imagen en escala de grises
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

# Verificar carga
if imagen is None:
    raise ValueError("No se pudo cargar la imagen. Verifica la ruta o el nombre del archivo.")

# Escalar al 150% (1.5x) y al 50% (0.5x)
imagen_escalada_150 = cv2.resize(imagen, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
imagen_escalada_50 = cv2.resize(imagen, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)

# Mostrar resultados
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagen, cmap='gray')
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagen_escalada_150, cmap='gray')
plt.title("Escala 150%")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagen_escalada_50, cmap='gray')
plt.title("Escala 50%")
plt.axis("off")

plt.tight_layout()
plt.show()
```

***Parte II.- Ecualización de histograma ***

Ejercicio 1: Ecualización.
*  Cargar una imagen médica con bajo contraste.
*  Obtener y visualizar el histograma de la imagen original.
*  Aplicar ecualización de histograma para mejorar el contraste.
*  Mostrar la imagen original, la imagen mejorada y comparar los histogramas.


En este ejercicio, vamos a mejorar el contraste de la imagen utilizando ecualización de histograma. Esto ajusta la distribución de intensidad de la imagen para mejorar los detalles en áreas oscuras y brillantes.
Antes de realizar la ecualización, mostramos el histograma de la imagen original para tener una referencia del contraste actual.
Aplicamos la ecualización de histograma para mejorar el contraste de la imagen.
Finalmente, comparamos los histogramas antes y después de la ecualización para ver cómo ha cambiado la distribución de los valores de intensidad.

```python
from google.colab import drive
drive.mount('/content/drive')

# Librerías necesarias
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ruta del archivo desde Drive
ruta_imagen = "/content/drive/My Drive/Imagen por Resonancia Magnética (IRM).png"

# Cargar imagen en escala de grises
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

# Verificar que la imagen se haya cargado
if imagen is None:
    raise ValueError("No se pudo cargar la imagen. Verifica la ruta o el nombre del archivo.")

# Aplicar ecualización de histograma
imagen_ecualizada = cv2.equalizeHist(imagen)

# Calcular histogramas
hist_original = cv2.calcHist([imagen], [0], None, [256], [0, 256])
hist_ecualizada = cv2.calcHist([imagen_ecualizada], [0], None, [256], [0, 256])

# Mostrar imágenes y histogramas
plt.figure(figsize=(12, 8))

# Imagen original
plt.subplot(2, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title("Imagen Original")
plt.axis("off")

# Histograma original
plt.subplot(2, 2, 2)
plt.plot(hist_original, color='black')
plt.title("Histograma Original")
plt.xlim([0, 256])

# Imagen ecualizada
plt.subplot(2, 2, 3)
plt.imshow(imagen_ecualizada, cmap='gray')
plt.title("Imagen Ecualizada")
plt.axis("off")

# Histograma ecualizado
plt.subplot(2, 2, 4)
plt.plot(hist_ecualizada, color='black')
plt.title("Histograma Ecualizado")
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
